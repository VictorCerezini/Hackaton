```python
# ==========================================
# INSTALAÇÃO
# ==========================================
!pip install google-generativeai --quiet

# ==========================================
# IMPORTS
# ==========================================
import google.generativeai as genai
import os
import json
import re
from google.colab import files
from PIL import Image
import getpass

# ==========================================
# CONFIGURAÇÃO DA API KEY
# ==========================================
api_key = getpass.getpass("AIzaSyDneU4pV1GcHO32dKX4lGlP-6YUUQ6YUqI")
genai.configure(api_key=api_key)

# Modelo compatível com sua conta
model = genai.GenerativeModel("gemini-2.5-flash")

# ==========================================
# UPLOAD DA IMAGEM
# ==========================================
print("Faça upload do diagrama de arquitetura:")
uploaded = files.upload()
image_path = list(uploaded.keys())[0]
image = Image.open(image_path)

# ==========================================
# 1️⃣ EXTRAÇÃO DE COMPONENTES
# ==========================================
prompt = """
Analise o diagrama de arquitetura e realize uma modelagem de ameaças STRIDE completa.
Para cada componente e conexão identificada, analise:
1. S (Spoofing)
2. T (Tampering)
3. R (Repudiation)
4. I (Information Disclosure)
5. D (Denial of Service)
6. E (Elevation of Privilege)

Retorne um JSON com esta estrutura:
{
  "analysis": [
    {
      "component": "Nome",
      "threats": [
        {"category": "STRIDE Category", "risk": "Descrição do risco", "mitigation": "Sugestão de correção"}
      ]
    }
  ]
}
Responda apenas JSON puro.
"""

response = model.generate_content([prompt, image])
text_response = response.text.strip()

# ==========================================
# EXTRAÇÃO ROBUSTA DO JSON (CORREÇÃO)
# ==========================================
json_match = re.search(r'\{.*\}', text_response, re.DOTALL)

if json_match:
    json_string = json_match.group()
    components_data = json.loads(json_string)
else:
    raise ValueError("Nenhum JSON válido encontrado na resposta do modelo.")

print("\n📦 COMPONENTES IDENTIFICADOS:\n")
print(json.dumps(components_data, indent=2, ensure_ascii=False))

# ==========================================
# 2️⃣ ENGINE STRIDE
# ==========================================
stride_map = {
    "User": ["Spoofing"],
    "API": ["Tampering", "Information Disclosure", "Denial of Service"],
    "Server": ["Tampering", "Denial of Service", "Elevation of Privilege"],
    "Database": ["Information Disclosure", "Tampering", "Denial of Service"],
    "External System": ["Spoofing", "Tampering"]
}

vulnerability_knowledge_base = {
    "Spoofing": {
        "risk": "Falsificação de identidade",
        "mitigation": "Autenticação forte e MFA"
    },
    "Tampering": {
        "risk": "Manipulação de dados",
        "mitigation": "Assinatura digital e validação de integridade"
    },
    "Information Disclosure": {
        "risk": "Vazamento de dados sensíveis",
        "mitigation": "Criptografia e controle de acesso"
    },
    "Denial of Service": {
        "risk": "Indisponibilidade do sistema",
        "mitigation": "Rate limiting e monitoramento"
    },
    "Elevation of Privilege": {
        "risk": "Acesso não autorizado a recursos críticos",
        "mitigation": "Controle de acesso baseado em papéis (RBAC)"
    }
}

# ==========================================
# 3️⃣ EXIBIÇÃO DO RELATÓRIO DE AMEAÇAS (IA GENERATIVA)
# ==========================================
print("\n🔐 RELATÓRIO DE MODELAGEM DE AMEAÇAS (STRIDE) - GERADO POR IA:\n")
print("=" * 80)

# O novo JSON usa "analysis" em vez de "components"
if "analysis" in components_data:
    for item in components_data["analysis"]:
        componente = item.get("component", "N/A")
        print(f"\n🏗️ COMPONENTE: {componente.upper()}")
        print("-" * 40)

        for threat in item.get("threats", []):
            print(f"🔹 Categoria: {threat.get('category')}")
            print(f"   🔴 Risco: {threat.get('risk')}")
            print(f"   ✅ Mitigação: {threat.get('mitigation')}")
            print("   " + "." * 30)
else:
    print("Erro: O formato do JSON retornado não contém a chave 'analysis'.")
    print("Resposta recebida:", json.dumps(components_data, indent=2))

print("\n" + "=" * 80)
print("FIM DO RELATÓRIO")
```
